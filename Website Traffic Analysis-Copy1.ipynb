```python
df.drop_duplicates(inplace=True)
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='event', data=df)
plt.title("Distribution of Events")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))

df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))

df['city'].value_counts().head(10).plot(kind='bar', color='orange')

plt.title("Top 10 Cities")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,5))

df['artist'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Events")
plt.show()
```

```python
plt.figure(figsize=(12,5))

df['album'].value_counts().head(10).plot(kind='bar', color='green')

plt.title("Top 10 Albums")
plt.xlabel("Album")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,5))

df['track'].value_counts().head(10).plot(kind='bar', color='purple')

plt.title("Top 10 Tracks")
plt.xlabel("Track")
plt.ylabel("Count")
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])

daily = df.groupby('date').size()

plt.figure(figsize=(15,5))

daily.plot()

plt.title("Events Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Events")
plt.show()
```

```python
plt.figure(figsize=(12,5))

df['linkid'].value_counts().head(10).plot(kind='bar', color='red')

plt.title("Top 10 Link IDs")
plt.xlabel("Link ID")
plt.ylabel("Count")
plt.show()
```

```python
df['event'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(7,7)
)

plt.title("Event Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,6))

df['country'].value_counts().head(10).plot(kind='barh')

plt.title("Top Countries")
plt.xlabel("Count")
plt.show()
```

```python
artist_event = df.groupby('artist')['event'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))

artist_event.plot(kind='bar')

plt.title("Top Artists by Event Count")
plt.xlabel("Artist")
plt.ylabel("Events")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values")
plt.show()
```
